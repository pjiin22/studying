- **matplotlib** \
: 파이썬 데이터 시각화의 가장 기본적인 라이브러리. 유연하게 커스텀이 가능하나 코드가 길고 어렵다는 단점이 있습니다.
- **seaborn**\
: matplotblib을 좀 더 쉽고 아름답게 시각화하는 라이브러리
- 실제 분석에서는 seaborn과 matplotlib을 함께 사용하여 시각화하는 경우가 많음

## **[1] 기본 문법**
```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(가로길이, 세로길이)) #그래프 크기 설정
sns.000plot(data=데이터, x=X축 컬럼, y=Y축 컬럼, hue=범례 컬럼) #seaborn으로 그래프 그리기
plt.title(제목) #제목 설정
plt.xlabel(라벨) #x축 라벨 설정
plt.ylabel(라벨) #y축 라벨 설정
plt.legend(loc=범례 위치 설정)
plt.xticks(rotation=x축 각도 설정)
plt.yticks(rotation=y축 각도 설정)
plt.show() #그래프 출력
```



````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = sns.load_dataset('penguins') #파이썬 제공 데이터셋
df
````

- barplot

````python
sns.barplot(data=df, x='species', y='body_mass_g')
````

````python
plt.figure(figsize=(10,5))
sns.barplot(data=df, x='species', y='body_mass_g')
plt.title('body mass per species')
plt.xlabel('per species')
plt.ylabel('body mass')
plt.show()
````

````python
sns.barplot(data=df, x="island", y="body_mass_g", hue="sex") #범례 추가
plt.show()
````

````python
sns.barplot(data=df, x="island", y="body_mass_g", hue="sex")
plt.xticks(rotation=30)
plt.legend(loc='upper left') #범례 위치 조정
plt.show()
````

## **[2] 스타일 설정하기**

```
sns.set_style(스타일)
sns.set_palette(팔레트)
```
- 스타일: darkgrid, whitegrid, dark, white
- 팔레트: [참고](https://seaborn.pydata.org/tutorial/color_palettes.html)


````python
sns.set_style('darkgrid')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_style('whitegrid')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_style('white')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_style('dark')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_palette('Set3')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_style('white')
sns.set_palette('flare') #그래프 스타일
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````

````python
sns.set_palette('crest')
sns.barplot(data=df, x='species', y='body_mass_g')
plt.show()
````
