### **loc, iloc 사용하기**

- loc: 레이블값을 사용하여 조회

```python
df.loc[3,]
#3번 인덱스 행을 선택
#즉 컬럼명 passengerid의 값은 4인 것
```

## **[2] 원하는 조건으로 데이터 추출하기**

### **데이터 정렬하기**

- 내림차순으로 정렬하고 싶다면 `ascending=False` 조건을 걸어주면 됩니다.

```python
df.head()
```

```python
df.sort_values('Age') #Age가 오름차순이 되도록
```

```python
df.sort_values('Age', ascending=False)
```

```python
df.sort_values(['Age','Fare'], ascending=[False,True])
```

### **특정 조건을 충족하는 데이터 추출하기**
- `데이터프레임명[조건식]`
- `데이터프레임명.query('조건식')`



```python
df[df['Pclass'] == 1] #꼭 두번 씌우기!
```

```python
df[(df['Pclass'] == 1) & (df['Age'] >= 30)] #조건이 두개이면 & 쓰기
```

```python
df[(df['Pclass'] == 1) | (df['Age'] >= 30)] #or
```

```python
df.query('Pclass == 1')
#query() 메소드를 사용하여 데이터프레임 df에서 'Pclass' 열이 1인 행들만 선택하는 코드
```

```python
df.query('Pclass == 1 and Age >= 30')
```

```python
df.query('Pclass == 1 or Age >= 30') #직접 or이라고 치기
```

```python
df[df['PassengerId'].isin([3,100,500])]
#or을 사용하면 같은 의미가 됨
#df[(df['PassengerId'] == 3) | (df['PassengerId'] == 100) | (df['PassengerId'] == 500)]
```

```python
df.query('PassengerId in [3,100,500]') #위와 같은 의미
```

```python
df.query('PassengerId.isin([3,100,500])') #경우에 따라 query 사용하기도 함
```

```python
import random

passengerid_list = list(df['PassengerId'])
passengerid_sample = random.sample(passengerid_list, 10)
```

```python
df[df['PassengerId'].isin(passengerid_sample)] #랜덤하게 추출
```

```python
df.query('PassengerId in @passengerid_sample')
# query() 메소드에서 사용되는 @ 기호는 파이썬 변수를 참조하기 위해 사용
# df: 데이터프레임을 의미합니다.
# .query(): 문자열로 조건을 걸어 데이터를 필터링하는 함수입니다.
# 'PassengerId in @passengerid_sample': 데이터프레임의 "PassengerId" 컬럼에서 passengerid_sample 리스트에 있는 값들만 선택하는 조건입니다.
# PassengerId: 데이터프레임의 컬럼 이름.
# in: "PassengerId" 값이 passengerid_sample 리스트 안에 있는지 확인.
# @passengerid_sample: 현재 변수로 존재하는 passengerid_sample 리스트를 참조하는 것을 의미합니다.
```
