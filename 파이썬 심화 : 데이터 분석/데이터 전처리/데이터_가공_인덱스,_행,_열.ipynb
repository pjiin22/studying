### **인덱스 변경**
- 인덱스 몇개를 바꿀 때
    - `데이터명.rename({인덱스:바꿀 인덱스, 인덱스:바꿀 인덱스, ...})`
- 인덱스 전체를 바꿀 때
    - `데이터명.index = 바꿀 인덱스 리스트`

```python
df1 = df.copy()
df1.rename({0:'row1', 1:'row2'})
```

```python
df1 = df1.rename({0:'row1', 1:'row2'}) #rename은 데이터프레임을 바꾸어주지 않으므로 변경을 하려면 다시 덧씌워주어야 합니다.
df1.head()
```

```python
df1.index = [i+1 for i in range(len(df1))]
df1.head()
```

### **열을 인덱스로 설정**
- `데이터명.set_index(컬럼명)`

```python
df1.set_index('PassengerId') #이 열이 행이 되도록 하겠다
```

```python
df1 = df1.set_index('PassengerId') #마찬가지로 변경을 하려면 덧씌워주어야 합니다.
df1.head()
```

### **인덱스를 열로 변환**
- 인덱스를 열로 변환하고 그 열을 남기고 싶으면
    - `데이터명.reset_index()`
- 인덱스를 열로 변환하고 그 열을 삭제하고 싶으면
    - `데이터명.reset_index(drop = True)`

```python
df1.reset_index() #앞서 행은 passengerid임 -> 현재 열이 됨
```

```python
df1.reset_index(drop=True) #행렬 변경 그리고 1열 날리기
```

```python
df1 = df1.reset_index() #함수에 다시 집어넣기
df1.head()
```

## **[2] 행**

### **행의 추가/제거**
- 행의 추가
    - `pd.concat([기존 데이터명, 붙일 데이터명])`
- 행의 제거
    - `데이터명.drop(인덱스명, axis=0)`

```python
df1 = df.copy()
df2 = df.copy()
```

```python
df3 = pd.concat([df1, df2])
df3 #똑같은거 두개를 연결시킴

#제일 끝 인덱스의 번호가 1781이 아니라 890임 즉 인덱스 그대로 이어붙여짐
```

```python
df3 = df3.reset_index(drop=True)
df3

#기존의 인덱스를 버리고 새로 0 ~ 1781 인덱스가 생김
```

```python
df3.drop([i for i in range(891, len(df3))])
# 인덱스 891부터 df3의 길이만큼(891개)을 제거[drop]
# 재정의 안함
```

## **[3] 열**

### **열의 추가/제거**

- 열의 추가
    - `데이터명[추가할 컬럼명] = 추가할 값`
- 열의 제거
    - `데이터명.drop(제거할 컬럼명, axis=1)`


> 컬럼은 []이고 인덱스는 index임




```python
df1 = df.copy()
df1['age_simplified'] = df1['Age']//10 * 10
```

```python
print(list(df1['Name'].str.split(','))) #문자열 컬럼에 str.split()을 쓰면 특정 문자열을 기준으로 자를 수 있습니다.
```

```python
df1['given_name'] = [i[0] for i in df1['Name'].str.split(',')]
df1.head()
```

```python
df1['string_add'] = df1['Sex']+df1['given_name']
df1.head()
```

```python
df1['number_add'] = df1['Age'] + df1['age_simplified']
df1.head()
```

```python
df1.drop('given_name', axis=1)
```

    - `데이터명.columns = 열 이름 리스트`

```python
df1 = df.copy()
df1
```

```python
# 열 이름인지 티내기 위해서 'axis = 1'하기
df1.rename({'PassengerId':'Id'}, axis=1)
```
