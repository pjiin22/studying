- `usecols`: 사용할 컬럼

```python
excel_data = pd.read_excel('/content/titanic_train.xlsx', sheet_name='시트1', header=1)
excel_data.head()
```

## **[3] 웹 html 파일**
- `pd.read_html(html경로)`
- `encoding`: 한글이 깨져서 나올 때 `utf-8/cp949`로 설정합니다.
